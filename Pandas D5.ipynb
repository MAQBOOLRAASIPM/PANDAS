```python
import pandas as pd

data = {
    "Name": ["John", "Sara", "Ali", "Maya", "Riya"],
    "Age": [16, 17, 15, 16, 18],
    "Marks": [45, 67, 89, 72, 91]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
df.sort_values("Age")
```

```python
df.describe()
```

```python
print(df.sum(numeric_only=True))
print(df["Marks"].sum())
print(df["Age"].max())
```

```python
import pandas as pd

data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department": ["HR", "IT", "Finance", "IT", "HR"],
    "Salary": [50000, 63000, 70000, 65000, 58000]
}

df = pd.DataFrame(data)
print(df)
```

```python

df.rename(columns={"Name":"EmpName"})
```

```python
df["Salary"].mean()
df["Salary"].max()
df.sort_values(by=["Department","Salary"],ascending =[True,False])
```
