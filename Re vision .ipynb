```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Amit", "Riya", "John", "Sara", "Karan"],
    "Age": [25, np.nan, 22, 28, None],
    "Marks": [85, 92, None, 88, 76]
}

df = pd.DataFrame(data)
print(df)
```

```python
print(df["Marks"].sum())    # Sum of Marks
print(df["Marks"].mean())   # Average Marks
print(df["Marks"].median()) # Median
print(df["Marks"].mode())   # Mode
print(df["Marks"].std())    # Standard deviation
print(df["Marks"].var())    # Variance
print(df["Marks"].min())    # Minimum
print(df["Marks"].max())    # Maximum
print(df["Marks"].cumsum())   # Cumulative sum
print(df["Marks"].cumprod())  # Cumulative product
```
